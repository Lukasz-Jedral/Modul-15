```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
plt.plot(salary_values)
```

```python
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in PLN")
plt.bar(names, salary_values)
```

```python
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in PLN")
plt.barh(names, salary_values, color= ["green"])
```
